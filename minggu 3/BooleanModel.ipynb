```python
import os
import re
import glob
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# --- Download resource nltk kalau belum ada ---
nltk.download('punkt')
nltk.download('stopwords')

# --- Folder tempat file teks berada ---
folder_path = "data/"  # ubah sesuai nama foldermu

# --- Fungsi bantu: membersihkan teks ---
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)  # hapus karakter selain huruf
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in tokens if word not in stop_words]
    return filtered_words

# --- Membaca semua file teks di folder ---
print("üìÇ Membaca semua dokumen...")
inverted_index = {}
file_list = glob.glob(os.path.join(folder_path, "*.txt"))

for file_path in file_list:
    try:
        with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
            text = f.read()
    except:
        print(f"‚ö†Ô∏è Tidak bisa membaca file: {file_path}")
        continue

    words = clean_text(text)
    for word in set(words):
        if word not in inverted_index:
            inverted_index[word] = set()
        inverted_index[word].add(file_path)

print(f"‚úÖ Indexing selesai! Total kata unik: {len(inverted_index)}")
print(f"üìÑ Total file: {len(file_list)}")

# --- Fungsi pencarian boolean ---
def boolean_search(query):
    query = query.lower().strip()
    if not query:
        print("‚ö†Ô∏è Kamu belum mengetikkan query apa pun.")
        return

    tokens = query.split()
    operators = {"and", "or", "not"}
    search_terms = [t for t in tokens if t not in operators]

    print(f"üîó Operator logika: {[t for t in tokens if t in operators]}")
    print(f"üß© Kata pencarian: {search_terms}")

    if not search_terms:
        print("‚ùå Tidak ada kata pencarian yang valid.")
        return

    # Ambil dokumen yang mengandung kata pertama
    result = inverted_index.get(search_terms[0], set())

    i = 1
    while i < len(tokens):
        if tokens[i] == "and" and i + 1 < len(tokens):
            result = result & inverted_index.get(tokens[i+1], set())
        elif tokens[i] == "or" and i + 1 < len(tokens):
            result = result | inverted_index.get(tokens[i+1], set())
        elif tokens[i] == "not" and i + 1 < len(tokens):
            all_docs = set(file_list)
            result = result - inverted_index.get(tokens[i+1], set())
        i += 1

    if result:
        print("\n‚úÖ Dokumen yang cocok dengan query:")
        for doc in result:
            print("üìÑ", doc)
    else:
        print("\n‚ùå Tidak ada dokumen yang cocok dengan query.")

# --- Input query dari pengguna ---
query = input("\nüîç badminton")
boolean_search(query)
```
